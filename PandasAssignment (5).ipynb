````python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
````

Q1: Print the shape of the dataset

````python
df = pd.read_csv("travel-times.csv")

df.shape
````

Q2: Print the first 15 rows of the dataset

````python
df.head(15)
````

Q3: Get information for the features (columns) with missing values

````python
df.info()
````

Q4: Drop duplicate values (if any) by keeping only the last instance

````python
df.drop_duplicates(inplace=True, keep='last')
````

Q5: Calculate the total number of missing values (if any) on each column

````python
df.isnull().sum()
````

Q6:
- Create two copies of the dataframe, and then:
    - drop rows with missing values from the 1st copy
    - drop columns with missing values from the 2nd copy

````python
df_copy1 = df.copy()
df_copy2 = df.copy()

df_copy1 = df_copy1.dropna()

df_copy2 = df_copy2.dropna(axis=1)
````

Q9: Create a subset with trips occurred on November 23, 2011 and January 6, 2012

````python
df['Date'] = pd.to_datetime(df['Date'])

subset_df = df[(df['Date'] == "2011-11-23") | (df['Date'] == "2012-01-06")]

subset_df
````

Q10: Produce a scatterplot between Distance and TotalTime

````python
plt.figure(figsize=(10, 8))
plt.scatter(df['Distance'], df['TotalTime'], alpha=0.5)
plt.xlabel('Distance')
plt.ylabel('Total Time')
plt.title('Scatterplot of Distance vs. Total Time')
plt.grid(True)
plt.show()
````

Q11: Produce boxplots for AvgSpeed and AvgMovingSpeed (use different cells for each)

````python
plt.figure(figsize=(8, 6))
plt.boxplot(df['AvgSpeed'], vert=True)
plt.xlabel("AvgSpeed")
plt.title("Boxplot of AvgSpeed")
plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(8, 6))
plt.boxplot(df['AvgMovingSpeed'], vert=True)
plt.xlabel("AvgMovingSpeed")
plt.title("Boxplot of AvgMovingSpeed")
plt.grid(True)
plt.show()
````
